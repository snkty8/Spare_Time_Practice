```python
import pandas as pd
```

```python
df1 = pd.read_csv(('Cleaned_Kaggle_2018.csv'))
df2 = pd.read_csv(('Cleaned_Kaggle_2019.csv'))
df3 = pd.read_csv(('Cleaned_Kaggle_2020.csv'))
```

```python
frames = [df1, df2, df3]
```

```python
df = pd.concat(frames)
df
```

```python
male = []
for x in df['Gender']:
    male.append(x.replace('Man','Male'))
df['Gender'] = male
```

```python
Female = []
for x in df['Gender']:
    Female.append(x.replace('Woman','Female'))
df['Gender'] = Female
```

```python
df
```

```python
df.to_csv(r'Cleaned_Combined_Kaggle_Data.csv', index = False)
```
